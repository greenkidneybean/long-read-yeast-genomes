```python
import pandas as pd
import numpy as np
```

```python
#ref = 's288c-orf'
genes = ['MEL1','XDH1','MAL11','MAL12','MAL13','MAL31','MAL32','MAL33','OPT2','SGE1','SUC2']
genomes = ['MSY24','MSY25','MSY26','MSY27','MSY28','MSY29','MSY30','MSY31','MSY32','MSY33','MSY34','MSY35','MSY36','MSY37','MSY38','MSY39']
blast_cols=['query','hit','percent_id','length','mismatch','gapopen','qstart','qend','tstart','tend','evalue','bitscore']
df_cols = ['strain','header','s288c_hit','percent_id']
```

```python
# iterate through each df and add 
df_list = []

for gene in genes:
    for genome in genomes:
        df = pd.read_csv(f'../data/blast_gene-hits/{gene}/{genome}.tsv', sep='\t', names=blast_cols)
        df['hit_strain'] = genome
        df_list.append(df)
df = pd.concat(df_list)
df.reset_index(inplace=True, drop=True)

df['hit_start'] = df.hit.str.split(':').str[1]
df['hit_end'] = df.hit.str.split(':').str[2].str.split(' ').str[0]
df['hit_name'] = df.hit.str.split(r"_").str[0].str[4:]
#df['hit_name'] = df.hit.str.split('_').str[1] + "_" + df['hit_name']
df['hit_chr'] = df.hit.str.split('_').str[2].str.split(':').str[0]
df['hit_strand'] = np.where(df.hit_start < df.hit_end, '+','-')
```
